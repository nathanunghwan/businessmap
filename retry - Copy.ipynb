```python
import pandas as pd
import folium
```

```python
# CSV 파일 불러오기
csv_path = 'korean_business_CSV.csv'
df = pd.read_csv(csv_path)

# 데이터 확인
df.head()
```

```python
desired_columns = ['Company Name', 'Executive First Name','Executive Last Name', 'Address', 'City','State','Phone Number Combined','Metro Area','Primary SIC Description','SIC Code 1 Description','SIC Code 2 Description','Location Sales Volume Actual']
new_df = df[desired_columns]
# 'Metro Area' 컬럼에서 데이터를 쉼표로 분리하여 새로운 컬럼 생성
new_df['Metro Area Split'] = df['Metro Area'].str.split(',')

# 분리된 데이터 중 첫 번째 요소를 'Metro Area 1' 컬럼으로 추가
new_df['Metro Area 1'] = new_df['Metro Area Split'].apply(lambda x: x[0] if isinstance(x, list) and len(x) > 0 else None)


new_df.drop('Metro Area Split', axis=1, inplace=True)
new_df.drop('Metro Area', axis=1, inplace=True)
# 'Address' 컬럼을 이용하여 'Address1' 컬럼 생성
new_df['Address1'] = new_df['Address'].str.split('#').str[0].str.strip()


# 결과 확인
new_df.to_csv('new_business_data.csv', index=False)
new_df.head()
```

```python
# 제거할 상태 코드 목록
remove_states = ['SC', 'TN', 'GA', 'AL', 'FL']

# 'State' 컬럼에서 값이 remove_states에 포함되지 않은 행들만 추출하여 new_df_filtered 생성
new_df_filtered = new_df[~new_df['State'].isin(remove_states)].copy()

# 결과 확인
new_df_filtered.head()
```

```python
# State 컬럼이 'TN'인 행들만 추출하여 TN_df 생성
TN_df = new_df[new_df['State'] == 'TN'].copy()
# 'TN_df' 데이터프레임의 인덱스 재설정
TN_df.reset_index(drop=True, inplace=True)
# 결과 확인
TN_df.head()
```

```python
from geopy.geocoders import Nominatim
from geopy.exc import GeocoderTimedOut

geolocator = Nominatim(user_agent='my_app')

# 위도(Latitude)와 경도(Longitude)를 저장할 빈 리스트 생성
latitudes = []
longitudes = []

# 'Address', 'City', 'State' 컬럼을 사용하여 위도와 경도 추출
for _, row in new_df.iterrows():
    address = row['Address']
    city = row['City']
    state = row['State']
    
    location = None
    try:
        location = geolocator.geocode(f"{address}, {city}, {state}", timeout=10)
    except GeocoderTimedOut:
        # 타임아웃 시간 초과 시 재시도
        location = geolocator.geocode(f"{address}, {city}, {state}", timeout=10)
    
    # 주소로부터 위도와 경도 추출하여 리스트에 추가
    if location is not None:
        latitudes.append(location.latitude)
        longitudes.append(location.longitude)
    else:
        latitudes.append(None)
        longitudes.append(None)

# '위도(Latitude)'와 '경도(Longitude)' 컬럼 추가
new_df['위도(Latitude)'] = latitudes
new_df['경도(Longitude)'] = longitudes

# 결과 확인
new_df.head()
```
